```python
yuv_path = 'test.yuv'
alpha_candidates = [0.5, 1.0, 1.5, 2.0]
bitrates = [1000, 2000, 3000]
for alpha in alpha_candidates:
    for bitrate in bitrates:
        encode_single_frame_yuv(yuv_path, alpha, bitrate, 'example_output/alpha_{}_bitrate_{}.mp4'.format(alpha, bitrate))
```

```python
encoded_dir = 'example_output'

jobs = []
for alpha in alpha_candidates:
    for bitrate in bitrates:
        encoded_path = os.path.join(encoded_dir, f'alpha_{alpha}_bitrate_{bitrate}.mp4')
        if os.path.isfile(encoded_path):
            jobs.append((encoded_path, alpha, bitrate))
jobs
```

```python
width = 1920
height = 1080
alpha_scores = {a: {} for a in alpha_candidates}
for encoded_path, alpha, bitrate in jobs:
    video_paths = [(encoded_path, 1)]
    score_dicts = evaluate_clip_iqa_batch(video_paths, width, height, batch_size=1)
    alpha_scores[alpha][bitrate] = score_dicts[0]
```
